```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# %%
# Define the file path for the dataset
file_path = 'CRASH_22_23_24.csv'

# Load the dataset from the CSV file into a DataFrame called 'df'
df = pd.read_csv(file_path)

# Print the shape of the DataFrame to see how many rows and columns it has
print(f"Dataset loaded successfully with shape: {df.shape}")
```

```python
# %%
# Display the first 5 rows of the DataFrame to inspect the data
df.head()
```

```python
# %%
# Get a summary of the DataFrame including column names, non-null counts, and data types
df.info()
```

```python
# Step 3: Basic overview
print("Dataset information:")
display(df.info())

print("\nSummary statistics:")
display(df.describe(include='all').T)
```

```python
# %%
# Step 4: Check for missing values
missing = df.isna().sum().sort_values(ascending=False)
missing_percent = (missing / len(df)) * 100
missing_summary = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_percent})
display(missing_summary.head(20))

# %%
# Step 5: Drop irrelevant columns
# Use your data dictionary to adjust this list.
# Example: Suppose columns like 'Unnamed: 0', 'Crash_ID', 'Report_Number' aren't useful for EDA.
ignore_cols = ['Unnamed: 0', 'Report_Number']
df = df.drop(columns=[col for col in ignore_cols if col in df.columns], errors='ignore')

print("Dropped irrelevant columns. New shape:", df.shape)

# %%
```

```python
# Step 6: Check data types and unique values
types = df.dtypes.value_counts()
print("Data type counts:\n", types)

# For each categorical column, print number of unique values
cat_cols = [c for c in df.columns if df[c].dtype == 'object']
cat_summary = pd.DataFrame({
    'Unique Values': [df[c].nunique() for c in cat_cols],
    'Sample Values': [df[c].dropna().unique()[:5] for c in cat_cols]
}, index=cat_cols)
display(cat_summary.head(10))

# %%
```

```python
# Step 7: Numeric columns summary
num_cols = df.select_dtypes(include=[np.number]).columns
display(df[num_cols].describe().T)
```

```python
# Step 8: Correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[num_cols].corr(), annot=False, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap (Numeric Columns)")
plt.show()
```

```python
# Step 9: Univariate analysis — numeric columns
for col in num_cols[:5]:  # adjust as needed
    plt.figure(figsize=(6, 3))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Step 10: Univariate analysis — categorical columns
for col in cat_cols[:5]:  # adjust as needed
    plt.figure(figsize=(6, 3))
    df[col].value_counts().head(10).plot(kind='bar')
    plt.title(f"Top categories in {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
# %%
# Step: Clean LATITUDE format and analyze its relationship with FATAL_COUNT

def dms_to_decimal(dms_str):
    """
    Convert latitude strings like '40 18:27.925' into decimal degrees.
    Assumes format: degrees minutes:seconds.decimal
    """
    try:
        parts = dms_str.split()
        deg = float(parts[0])
        min_sec = parts[1].split(':')
        minutes = float(min_sec[0])
        seconds = float(min_sec[1]) if len(min_sec) > 1 else 0.0
        return deg + (minutes / 60) + (seconds / 3600)
    except Exception:
        return np.nan  # if format unexpected

# Convert LATITUDE if it's not already numeric
if df["LATITUDE"].dtype == "object":
    df["LATITUDE_DEC"] = df["LATITUDE"].apply(dms_to_decimal)
else:
    df["LATITUDE_DEC"] = df["LATITUDE"]

# Drop rows with missing values
subset = df[["LATITUDE_DEC", "FATAL_COUNT"]].dropna()

# Convert FATAL_COUNT to numeric if needed
subset["FATAL_COUNT"] = pd.to_numeric(subset["FATAL_COUNT"], errors="coerce")

# Remove rows where conversion failed
subset = subset.dropna()

# Compute correlation
corr_value = subset["LATITUDE_DEC"].corr(subset["FATAL_COUNT"])
print(f"Correlation between LATITUDE (decimal) and FATAL_COUNT: {corr_value:.4f}")

# Scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=subset, x="LATITUDE_DEC", y="FATAL_COUNT", alpha=0.6)
plt.title("Relationship between Latitude (Decimal Degrees) and Fatal Count")
plt.xlabel("Latitude (Decimal Degrees)")
plt.ylabel("Fatal Count")
plt.show()

# Aggregate by latitude bins
subset["Latitude_Bin"] = pd.cut(subset["LATITUDE_DEC"], bins=10)
grouped = subset.groupby("Latitude_Bin")["FATAL_COUNT"].mean().reset_index()

plt.figure(figsize=(8, 4))
sns.barplot(data=grouped, x="Latitude_Bin", y="FATAL_COUNT", color="steelblue")
plt.xticks(rotation=45)
plt.title("Average Fatal Count by Latitude Range")
plt.xlabel("Latitude Range")
plt.ylabel("Average Fatal Count")
plt.show()
```
